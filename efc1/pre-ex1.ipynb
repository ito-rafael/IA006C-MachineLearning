```python
#----------------------------
# import libraries
#----------------------------
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
#-----------------------------
# manipulating data
#-----------------------------
# read csv file
df = pd.read_csv('daily-minimum-temperatures.csv')
date = df['Date'].values
temp = df['Temp'].values
size = len(date)
#------------------
# input data
x_raw = temp[:-365]
y = np.array([temp]).T
y = y[:-365]
#------------------
# path to save figures
path = 'figures/pre-ex1/'
```

```python
# plot
plt.plot(temp, '')
# axis label
plt.xlabel(r'$N_{th}$ day')
plt.ylabel('Temperature [°C]')
# axis adjust
plt.autoscale(axis='x', tight=True)
# title
plt.title('Minimum daily temperature')
# add grid
plt.grid(True)
plt.savefig(path + 'figure_0_all_data.png', dpi=300)
plt.show()
```

```python
# plot
plt.plot(range(0*365,1*365), temp[0*365:1*365], 'b')
plt.plot(range(1*365,2*365), temp[1*365:2*365], 'g')
plt.plot(range(2*365,3*365), temp[2*365:3*365], 'r')
plt.plot(range(3*365,4*365), temp[3*365:4*365], 'c')
plt.plot(range(4*365,5*365), temp[4*365:5*365], 'm')
plt.plot(range(5*365,6*365), temp[5*365:6*365], 'y')
plt.plot(range(6*365,7*365), temp[6*365:7*365], 'k')
plt.plot(range(7*365,8*365), temp[7*365:8*365], 'violet')
plt.plot(range(8*365,9*365), temp[8*365:9*365], 'orange')
plt.plot(range(9*365,10*365), temp[9*365:10*365], 'navy')
# axis label
plt.xlabel(r'$N_{th}$ day')
plt.ylabel('Temperature [°C]')
# axis adjust
plt.autoscale(axis='x', tight=True)
# title
plt.title('Minimum daily temperature')
# add grid
plt.grid(True)
plt.savefig(path + 'figure_1_data_split.png', dpi=300)
plt.show()
```
