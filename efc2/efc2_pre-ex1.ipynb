```python
#----------------------------
# import libraries
#----------------------------
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# path to save figures
path = 'figures/pre-ex1/'
# read data
df = pd.read_csv('dados_voz_genero.csv')
# remove first column (index)
df = df.drop(df.columns[[0]], axis=1) 
# dataframe to numpy
features = df.to_numpy()
```

```python
for col in range(features.shape[1]):
    plt.title(df.columns[col])
    plt.hist(features[:,col], histtype='bar', ec='black')
    #plt.grid(True)
    plt.autoscale(axis='x', tight=True)
    plt.savefig(path + df.columns[col].replace('.','_'), dpi=300)
    plt.show()
```

```python
f = plt.figure(figsize=(19, 15))
plt.matshow(df.corr(), fignum=f.number)
plt.xticks(range(df.shape[1]), df.columns, fontsize=14, rotation=45)
plt.yticks(range(df.shape[1]), df.columns, fontsize=14)
cb = plt.colorbar()
cb.ax.tick_params(labelsize=14)
plt.title('Correlation Matrix', fontsize=16);
plt.savefig(path + 'correlation', dpi=300)
plt.show()
```

```python
x = np.linspace(-10, 10, 100)
y = 1 / (1 + np.exp(-x))
# plot
plt.plot(x, y, 'r')
# axis label
plt.xlabel('x')
plt.ylabel('g(x)')
# title
plt.title('Hyperbolic tangent')
# add grid
plt.grid(True)
plt.autoscale(axis='x', tight=True)
plt.savefig(path + 'logistic_function', dpi=300)
plt.show()
```

```python
x = np.linspace(-10, 10, 100)
y = 1 / (1 + np.exp(-x))
y = (2 * y) - 1
z = np.tanh(x)
# plot
plt.plot(x, y, 'r', label='logistic')
plt.plot(x, z, 'b', label='hyperbolic tangent')
# axis label
plt.xlabel('x')
plt.ylabel('g(x)')
# title
plt.title('Hyperbolic tangent')
# add grid
plt.grid(True)
plt.legend()
plt.autoscale(axis='x', tight=True)
plt.savefig(path + 'logistic_vs_tanh', dpi=300)
plt.show()
```
