```python
#----------------------------
# import libraries
#----------------------------
import numpy as np
import scipy.stats as stats
from matplotlib import pyplot as plt
#------------------
# path to save figures
path = 'figures/theoretical-ex/'
```

```python
# class 1 distribution
mu_1 = -1
variance_1 = 1
sigma_1 = np.sqrt(variance_1)
# class 2 distribution
mu_2 = 1
variance_2 = 1
sigma_2 = np.sqrt(variance_2)

# create x for classes C1 and C2
x1 = np.linspace(mu_1 - 3*sigma_1, mu_1 + 3*sigma_1, 100)
x2 = np.linspace(mu_2 - 3*sigma_2, mu_2 + 3*sigma_2, 100)
# create y for classes C1 and C2
y1 = stats.norm.pdf(x1, mu_1, sigma_1)
y2 = stats.norm.pdf(x2, mu_2, sigma_2)

# plot C1 and C2
plt.plot(x1, y1, label='p(x|C1)')
plt.plot(x2, y2, label='p(x|C2)')
# fill the graph 
plt.fill_between(x1, y1, color='#539ecd', alpha='1.0')
plt.fill_between(x2, y2, color='#ff8f53', alpha='1.0')

# axis label
plt.xlabel('x')
plt.ylabel('p(x)')
# title
plt.title('ML criteria')

plt.autoscale(axis='x', tight=True) # axis adjust
plt.grid(True) # add grid
plt.legend() # add legend
plt.savefig(path + 'ML.png', dpi=300)
plt.show()
```

```python
# class 1 distribution
mu_1 = -1
variance_1 = 1
sigma_1 = np.sqrt(variance_1)
# class 2 distribution
mu_2 = 1
variance_2 = 1
sigma_2 = np.sqrt(variance_2)

# create x for classes C1 and C2
x1 = np.linspace(mu_1 - 3*sigma_1, mu_1 + 3*sigma_1, 100)
x2 = np.linspace(mu_2 - 3*sigma_2, mu_2 + 3*sigma_2, 100)
# create y for classes C1 and C2
y1 = stats.norm.pdf(x1, mu_1, sigma_1)*0.7
y2 = stats.norm.pdf(x2, mu_2, sigma_2)*0.3

# plot C1 and C2
plt.plot(x1, y1, label='p(C1|x)')
plt.plot(x2, y2, label='p(C2|x)')
# fill the graph 
plt.fill_between(x1, y1, color='#539ecd', alpha='1.0')
plt.fill_between(x2, y2, color='#ff8f53', alpha='1.0')

# axis label
plt.xlabel('x')
plt.ylabel('p(x)')
# title
plt.title('MAP criteria')

plt.autoscale(axis='x', tight=True) # axis adjust
plt.grid(True) # add grid
plt.legend() # add legend
plt.savefig(path + 'MAP.png', dpi=300)
plt.show()
```
