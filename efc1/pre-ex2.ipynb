```python
#----------------------------
# import libraries
#----------------------------
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
def linear_transformation(x, lo, hi):
    # lo = lower limit desired
    # hi = higher limit desired
    # y = a * x + b
    max_x = np.max(x)
    min_x = np.min(x)
    # hi = a * max_x + b
    # lo = a * min_x + b
    a = (hi - lo) / (max_x - min_x)
    b = hi - (a * max_x)
    return a, b
```

```python
def feature_scaling(x, a, b):
    y = a * x + b
    return y
```

```python
#-----------------------------
# manipulating data
#-----------------------------
# read csv file
df = pd.read_csv('daily-minimum-temperatures.csv')
date = df['Date'].values
temp = df['Temp'].values
size = len(date)
#------------------
limits = [-1, 1]
a, b = linear_transformation(temp, limits[0], limits[1])
temp2 = feature_scaling(temp, a, b)
y = np.array([temp2]).T
#------------------
# path to save figures
path = 'figures/pre-ex2/'
```

```python
x = np.linspace(-1, 1, 100)
y = np.tanh(x)
# plot
plt.plot(x, y, 'g')
# axis label
plt.xlabel('x')
plt.ylabel('tanh(x)')
# title
plt.title('Hyperbolic tangent')
# add grid
plt.grid(True)
plt.autoscale(axis='x', tight=True)
plt.savefig(path + 'figure_1_hyperbolic_tangent', dpi=300)
plt.show()
```

```python
# plot
plt.plot(temp2, 'g')
# axis label
plt.xlabel(r'$N_{th}$ day')
plt.ylabel('Temperature linearly transformed')
# axis adjust
plt.autoscale(axis='x', tight=True)
# title
plt.title('Minimum daily temperature linearly transformed')
# add grid
plt.grid(True)
plt.savefig(path + 'figure_2_linear_transformation', dpi=300)
plt.show()
```
